# Visualizing the Backtest Calendar

 This short notebook gives two examples of how the backtest calendar is constructed: one using a monthly rebalancing frequency (Example 1), and one using a fixed number of trading days between rebalances (here, 10, in Example 2). All dates are snapped to the closest previous NYSE trading day.

 See §4 of the documentation for the full specification of `t_0`, `P`, `P_train`, and the backtest window.


```python
%load_ext autoreload
%autoreload 2
```

```python
import sys
from pathlib import Path

project_root = Path.cwd().parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from modules import backtest_calendar as bcal
```

### Example 1

```python
start = "2022-10-05"
freq_type = "monthly"
P = 8
P_train = 4


result = bcal.build_calendar(
    start = start,
    freq_type = freq_type,
    P = P,
    P_train = P_train
)


print("\n"+"=" * 60)
print("EXAMPLE 1: Monthly portfolio rebalancing")
print("=" * 60)
print(f"\nt_0 (aligned to closest previous NYSE trading date): {result.t0.date()}")
print(f"\nnumber of inter-rebalance periods: P = {P}")
print(f"\nRebalance dates [t_0, ... , t_P]:")
for i, d in enumerate(result.rebalance_dates):
    print(f"  t_{i:<3} {d.date()}")
print(f"\n\nnumber of training inter-rebalance periods: P_train = {P_train}")
print(f"\nTraining rebalance dates [t_{{-P_train}}, ... , t_0, ... , t_P]:")
offset = -len(result.training_rebalance_dates) + len(result.rebalance_dates)
for i, d in enumerate(result.training_rebalance_dates):
    label = i + offset
    print(f"  t_{label:<4} {d.date()}")
print(f"\n\nBacktest dates: {len(result.backtest_dates)} NYSE trading days "
      f"from {result.backtest_dates[0].date()} (t_0) to {result.backtest_dates[-1].date()} (one period after t_P)")
print("\n")
```

### Example 2

```python
start = "2022-10-05"
freq_type = "ndays"
freq = 14
P = 8
P_train = 4


result = bcal.build_calendar(
    start = start,
    freq_type = freq_type,
    P = P,
    P_train = P_train,
    freq = freq
)


print("\n" + "=" * 60)
print(f"EXAMPLE 2: Rebalancing the portfolio every {freq} trading days")
print("=" * 60)
print(f"\nt_0 (aligned to closest previous NYSE trading date): {result.t0.date()}")
print(f"\nnumber of inter-rebalance periods: P = {P}")
print(f"\nRebalance dates [t_0, ... , t_P]:")
for i, d in enumerate(result.rebalance_dates):
    print(f"  t_{i:<3} {d.date()}")
print(f"\n\nnumber of training inter-rebalance periods: P_train = {P_train}")
print(f"\nTraining rebalance dates [t_{{-P_train}}, ... , t_0, ... , t_P]:")
offset = -len(result.training_rebalance_dates) + len(result.rebalance_dates)
for i, d in enumerate(result.training_rebalance_dates):
    label = i + offset
    print(f"  t_{label:<4} {d.date()}")
print(f"\n\nBacktest dates: {len(result.backtest_dates)} NYSE trading days "
      f"from {result.backtest_dates[0].date()} (t_0) to {result.backtest_dates[-1].date()} (one period after t_P)")
print("\n")
```
